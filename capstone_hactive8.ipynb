## Analisis 1: **Pengaruh Gender Terhadap Skor**

```python
import pandas as pd

# Kode ini akan membaca file yang barusan kamu upload
df = pd.read_csv('StudentsPerformance.csv')

# Kode ini untuk menampilkan 5 baris pertama dari data untuk cek
df.head()
```

```python
# rata-rata skor matematika, membaca, dan menulis berdasarkan gender
gender_scores = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()

# nampilin hasil tabel rata-ratanya
print("Rata-rata Skor Berdasarkan Gender:")
print(gender_scores)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Membuat plot menggunakan data gender_scores
gender_scores.plot(kind='bar', figsize=(10, 6))

# Menambahkan judul dan label
plt.title('Perbandingan Rata-rata Skor Berdasarkan Gender', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Rata-rata Skor', fontsize=12)
plt.xticks(rotation=0) # Agar label 'female' dan 'male' tidak miring
plt.legend(title='Mata Pelajaran')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

## Analisis 2 : **Pengaruh Kursus Persiapan Ujian**

```python
# Menghitung rata-rata skor berdasarkan status kursus persiapan
prep_course_scores = df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean()

# Menampilkan hasil tabelnya
print("Rata-rata Skor Berdasarkan Kursus Persiapan:")
print(prep_course_scores)
```

```python
# Membuat visualisasi untuk perbandingan skor berdasarkan kursus persiapan
prep_course_scores.plot(kind='bar', figsize=(10, 6))

# Menambahkan judul dan label
plt.title('Pengaruh Kursus Persiapan Terhadap Skor Ujian', fontsize=16)
plt.xlabel('Status Kursus Persiapan', fontsize=12)
plt.ylabel('Rata-rata Skor', fontsize=12)
plt.xticks(rotation=0) # Agar label 'completed' dan 'none' tidak miring
plt.legend(title='Mata Pelajaran')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

## Analisis 3: **Pengaruh Tipe Makan Siang**

```python
# Menghitung rata-rata skor berdasarkan tipe makan siang
lunch_scores = df.groupby('lunch')[['math score', 'reading score', 'writing score']].mean()

# Menampilkan hasil tabelnya
print("Rata-rata Skor Berdasarkan Tipe Makan Siang:")
print(lunch_scores)
```

```python
# Membuat visualisasi untuk perbandingan skor berdasarkan tipe makan siang
lunch_scores.plot(kind='bar', figsize=(10, 6))

# Menambahkan judul dan label
plt.title('Pengaruh Tipe Makan Siang Terhadap Skor Ujian', fontsize=16)
plt.xlabel('Tipe Makan Siang', fontsize=12)
plt.ylabel('Rata-rata Skor', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Mata Pelajaran')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Menampilkan plot
plt.show()
```
